```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_yellowpages(search, location, pages=2):
    businesses = []

    for page in range(1, pages + 1):
        url = f"https://www.yellowpages.com/search?search_terms={search}&geo_location_terms={location}&page={page}"
        headers = {"User-Agent": "Mozilla/5.0"}

        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")

        results = soup.find_all("div", class_="result")

        for item in results:
            name = item.find("a", class_="business-name")
            phone = item.find("div", class_="phone")
            address = item.find("div", class_="street-address")
            website = item.find("a", class_="track-visit-website")

            businesses.append({
                "name": name.text.strip() if name else "",
                "phone": phone.text.strip() if phone else "",
                "address": address.text.strip() if address else "",
                "website": website.get("href", "") if website else ""
            })

        time.sleep(1)

    return pd.DataFrame(businesses)

df = scrape_yellowpages("restaurant", "New York", pages=2)
print(f"Found {len(df)} businesses")
df.head()
```

```python
df.to_csv("leadlist.csv", index=False)
print("Saved to leadlist.csv")
```
